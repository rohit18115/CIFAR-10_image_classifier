## Load the Dataset

Specify your transforms as a list first.
The transforms module is already loaded as `transforms`.

CIFAR-10 is fortunately included in the torchvision module.
Then, you can create your dataset using the `CIFAR10` object from `torchvision.datasets` ([the documentation is available here](https://pytorch.org/docs/stable/torchvision/datasets.html#cifar)).
Make sure to specify `download=True`! 

Once your dataset is created, you'll also need to define a `DataLoader` from the `torch.utils.data` module for both the train and the test set.

```python
data_dir = './data/CIFAR-10-images'
print(os.listdir(data_dir))
classes= os.listdir(os.path.join(data_dir,'train'))
print(classes)
```

```python
train_dict = {}
for i in classes:
    train_dict[i] = os.listdir(os.path.join(data_dir,'train',i))
    print('No of training example for {}:'.format(i),len(train_dict[i]))
```

```python
test_dict = {}
for i in classes:
    test_dict[i] = os.listdir(os.path.join(data_dir,'test',i)) 
    print('No of testing example for {}:'.format(i),len(test_dict[i]))
```

```python
random_seed=42
torch.manual_seed(random_seed);
```

```python
# Define transforms
train_transforms = transforms.Compose([transforms.RandomRotation(30),
                                       transforms.RandomResizedCrop(224),
                                       transforms.RandomHorizontalFlip(),
                                       transforms.ToTensor(),
                                       transforms.Normalize([0.485, 0.456, 0.406],
                                                            [0.229, 0.224, 0.225])])

test_transforms = transforms.Compose([transforms.Resize(255),
                                      transforms.CenterCrop(224),
                                      transforms.ToTensor(),
                                      transforms.Normalize([0.485, 0.456, 0.406],
                                                           [0.229, 0.224, 0.225])])
# Create training set and define training dataloader
# train and val dataset object made from ImageFolder class(WITHOUT TRANSFORM)

trainset = ImageFolder(data_dir+'/train', transform=transforms.ToTensor())
val_size = 5000
train_size = len(trainset) - val_size
trainset, valset = random_split(trainset,[train_size, val_size],generator=torch.manual_seed(random_seed))
len(trainset), len(valset)

# train and val dataset object made from ImageFolder class(WITh TRANSFORM)

trainset_transforms = ImageFolder(data_dir+'/train', transform=train_transforms)
train_size = len(trainset_transforms) - val_size
trainset_transforms, valset_transforms = random_split(trainset_transforms,[train_size, val_size],generator=torch.manual_seed(random_seed))
len(trainset_transforms), len(valset_transforms)
```

```python
# Dataloaders for the train and val sets for both with and without transforms
trainloader = torch.utils.data.DataLoader(trainset, batch_size=64, shuffle=True)
valloader = torch.utils.data.DataLoader(valset, batch_size=64, shuffle=True)
trainloader_transforms = torch.utils.data.DataLoader(trainset_transforms, batch_size=64, shuffle=True)
valloader_transforms = torch.utils.data.DataLoader(valset_transforms, batch_size=64, shuffle=True)
```

```python
# Create test set and define test dataloader

testset = ImageFolder(data_dir+'/test', transform=transforms.ToTensor())
testset_transforms = ImageFolder(data_dir+'/test', transform=test_transforms)
testloader = torch.utils.data.DataLoader(testset, batch_size=64)
testloader_transforms = torch.utils.data.DataLoader(testset_transforms, batch_size=64, shuffle=False)
```

```python

# Use GPU if it's available
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(device)

## YOUR CODE HERE ##
class Cifar10Model(nn.Module):
    def __init__(self):
        super().__init__()
        self.network = nn.Sequential(
            nn.Conv2d(3, 32, kernel_size=3, padding=1),
            nn.ReLU(),
            nn.Conv2d(32, 64, kernel_size=3, stride=1, padding=1),
            nn.ReLU(),
            nn.MaxPool2d(2, 2), # output: 64 x 16 x 16

            nn.Conv2d(64, 128, kernel_size=3, stride=1, padding=1),
            nn.ReLU(),
            nn.Conv2d(128, 128, kernel_size=3, stride=1, padding=1),
            nn.ReLU(),
            nn.MaxPool2d(2, 2), # output: 128 x 8 x 8

            nn.Conv2d(128, 128, kernel_size=3, stride=1, padding=1),
            nn.ReLU(),
#             nn.Conv2d(64, 64, kernel_size=3, stride=1, padding=1),
#             nn.ReLU(),
            nn.MaxPool2d(2, 2), # output: 128 x 4 x 4

            nn.Flatten(), 
            nn.Linear(2048, 1024),
            nn.ReLU(),
            nn.Dropout(0.1),
            nn.Linear(1024, 512),
            nn.ReLU(),
            nn.Dropout(0.1),
            nn.Linear(512, 10))
        
    def forward(self, xb):
        return self.network(xb)
model = Cifar10Model()

# Freeze parameters so we don't backprop through them

# for param in model.parameters():
#     param.requires_grad = False

# model.classifier = nn.Sequential(nn.Linear(25088, 2048),
#                                  nn.ReLU(),
#                                  nn.Linear(2048, 256),
#                                  nn.ReLU(),
#                                  nn.Linear(256, 10),
#                                  nn.LogSoftmax(dim=1))


criterion = nn.CrossEntropyLoss()

# Only train the classifier parameters, feature parameters are frozen
optimizer = optim.Adam(model.parameters(), lr=0.001)

conf = {
    'model_name': 'cnn',
    'group': 'scratch',
    # 'classifier_layers_sizes':[25088,2048],
    'num_classes':10,
    'loss': criterion,
    'input_dim': (3,32,32),
    'dropout': 0.1,
    'hidden_layer_activation':nn.ReLU(),
    'batch_size':64,
    'lr':0.001,
    'optimizer': optimizer
    
}

wandb.init(project = 'intro to machine learning nanodegree project 2', group = 'scratch', config = conf)

model.to(device);
model
```

## Running your Neural Network
Use whatever method you like to train your neural network, and ensure you record the average loss at each epoch. 
Don't forget to use `torch.device()` and the `.to()` method for both your model and your data if you are using GPU!

If you want to print your loss during each epoch, you can use the `enumerate` function and print the loss after a set number of batches. 250 batches works well for most people!

```python
epochs = 40
steps = 0

train_losses, val_losses = [], []
for e in range(epochs):
    running_loss = 0
    train_accuracy = 0
    for images, labels in tqdm_notebook(trainloader):
        images = images.to(device)
        labels = labels.to(device)
#         print(images.shape, labels.shape)
        optimizer.zero_grad()
        
        log_ps = model(images)
        loss = criterion(log_ps, labels)
        loss.backward()
        optimizer.step()
        
        running_loss += loss.item()
        ps = torch.exp(log_ps)
        top_p, top_class = ps.topk(1, dim=1)
        equals = top_class == labels.view(*top_class.shape)
        train_accuracy += torch.mean(equals.type(torch.FloatTensor))
        
        
    else:
        val_loss = 0
        val_accuracy = 0
        
        # Turn off gradients for validation, saves memory and computations
        with torch.no_grad():
            model.eval()
            for images, labels in tqdm_notebook(valloader):
                images = images.to(device)
                labels = labels.to(device)
                log_ps = model(images)
                val_loss += criterion(log_ps, labels)
                
                ps = torch.exp(log_ps)
                top_p, top_class = ps.topk(1, dim=1)
                equals = top_class == labels.view(*top_class.shape)
                val_accuracy += torch.mean(equals.type(torch.FloatTensor))
        
        model.train()
        
        
        train_losses.append(running_loss/len(trainloader))
        val_losses.append(val_loss/len(valloader))
        wandb.log({"train_loss":train_losses[-1], 
                   "val_loss":val_losses[-1], 
                   "train_accuracy": train_accuracy/len(trainloader),
                   "val_accuracy": val_accuracy/len(valloader)})
        print("Epoch: {}/{}.. ".format(e+1, epochs),
              "Training Loss: {:.3f}.. ".format(train_losses[-1]),
              "Val Loss: {:.3f}.. ".format(val_losses[-1]),
              "train Accuracy: {:.3f}".format(train_accuracy/len(trainloader)),
              "Val Accuracy: {:.3f}".format(val_accuracy/len(valloader)))
```

Plot the training loss (and validation loss/accuracy, if recorded).

```python
plt.plot(train_losses, label='Training loss')
plt.plot(val_losses, label='Validation loss')
plt.legend(frameon=False)
```

```python
def view_classify(img, ps, version="MNIST"):
    ''' Function for viewing an image and it's predicted classes.
    '''
    ps = ps.data.numpy().squeeze()

    fig, (ax1, ax2) = plt.subplots(figsize=(6,9), ncols=2)
    ax1.imshow(img.T)
    ax1.axis('off')
    ax2.barh(np.arange(10), ps)
    ax2.set_aspect(0.1)
    ax2.set_yticks(np.arange(10))
    if version == "MNIST":
        ax2.set_yticklabels(np.arange(10))
    elif version == "Fashion":
        ax2.set_yticklabels(['T-shirt/top',
                            'Trouser',
                            'Pullover',
                            'Dress',
                            'Coat',
                            'Sandal',
                            'Shirt',
                            'Sneaker',
                            'Bag',
                            'Ankle Boot'], size='small');
    if version == "CIFAR10":
        ax2.set_yticklabels(['airplane', 
                             'automobile', 
                             'bird', 
                             'cat', 
                             'deer', 
                             'dog', 
                             'frog', 
                             'horse', 
                             'ship', 
                             'truck'], size='small')
    ax2.set_title('Class Probability')
    ax2.set_xlim(0, 1.0)

    plt.tight_layout()
```

```python
# Test out your network!

model.eval()

dataiter = iter(testloader)
images, labels = dataiter.next()
img = images[50]
print(img.shape)
# Calculate the class probabilities (softmax) for img
with torch.no_grad():
    output = model.forward(images[50].resize_(1,3,32,32).to(device))

ps = output.cpu()

# Plot the image and probabilities
view_classify(img.view(3, 32, 32), ps, version='CIFAR10')
```
